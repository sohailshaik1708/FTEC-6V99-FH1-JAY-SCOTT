### Lesson 12 (Regular Expressions)

```python
import re

file = open('mbox.txt')
email_list = []
for line in file:
    line = line.rstrip()
    email_correspondence = re.findall('^From (\S+@\S+)',line)
    email_list.append(email_correspondence)
non_empty_list = [x for x in email_list if x]
print(non_empty_list)
```

### Question 1

```python
def annual_cashflow(rate_per,npv,time_period):
    cash_flow = rate_per * npv *0.01 / (1-(1+rate_per *0.01)** -time_period)
    return cash_flow
```

```python
annual_cashflow(6.25,31000,14)
```

### Question 3

```python
def discount_rate(time_period_in_months,pv,fv,frequency):
    n = time_period_in_months/frequency
    discount_rate = (fv/pv) ** (1/n)- 1
    return discount_rate
```

```python
discount_rate(72,1,3,3)
```

### Question 4

```python
def present_value_of_growing_annuity(payment, growth_rate, discount_rate, time_period):
    PV = payment * (1- ((1 + growth_rate)/ (1 + discount_rate))** time_period) / (discount_rate - growth_rate) 
    return PV
```

```python
present_value_of_growing_annuity(540000,0.05,0.10,16)
```

### Question 7

```python
from scipy.optimize import fsolve
print("Computing APR....")

def APR(r):
    n = 360

    pv = 2800000 * 0.8

    payment = 17000

    return payment - r * pv / (1 - (1 + r) ** (-n))

 

r = fsolve(APR, .01)
print('APR is {}'.format(r * 12))
```

```python
#calculation for EAR 

EAR = (1 + r)**12 - 1
EAR
```
